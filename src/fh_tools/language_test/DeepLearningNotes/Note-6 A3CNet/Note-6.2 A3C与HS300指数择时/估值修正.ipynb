```python
import numpy as np
import pandas as pd
from emulator.main import Account
```

```python
A = Account()
```

```python
actions = np.random.randint(0,3, size=(32))
```

```python
rewrads_list = []
for i in actions:
    r,n,d = A.step(i)
    rewrads_list.append(r)
```

```python
GAMMA = 0.9
R_list = []
R = 0
for i in rewrads_list[::-1]:
    R = i + GAMMA * R
    R_list.append(R)
R_list.reverse()
```

```python
tmp = pd.DataFrame()
tmp['reward'] = rewrads_list
tmp['discount'] = R_list
```
